```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    print(f'{i} kolonunun eksik verisi % {((df[i].isnull().sum())/(df[i].isnull().sum()+df[i].value_counts().sum()))*100}\n')
```

```python
plt.figure(figsize=(10,7),dpi=300)
sns.boxplot(df['age'],color='blue')
#OK FAZLA OUTLIER YOK SE MEAN VARSA MEDIAN KOYMAK DAHA MANTIKLI
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df.drop('deck',axis=1,inplace=True)
```

```python
df[df['embarked'].notna()]['embarked'].mode() #EN OK TEKRAR EDEN DEER ALDIK
```

```python
mode_value=df[df['embarked'].notna()]['embarked'].mode() 
```

```python
df['embarked_mode']=df['embarked'].fillna(mode_value)
```

```python
df[['embarked_mode','embarked']]
```

```python
df['embark_town']=df['embark_town'].fillna(df[df['embark_town'].notna()]['embark_town'].mode(),inplace=True)
```
