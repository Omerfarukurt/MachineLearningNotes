```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.datasets import make_classification
```

```python
#random stabilize
np.random.seed(42)
set1=900
set2=100
```

```python
df1=pd.DataFrame({
    'feature1': np.random.normal(loc=0,scale=1,size=set1),
    'feature2': np.random.normal(loc=0,scale=1,size=set1),
    'target': [0]*set1
})

df2=pd.DataFrame({
        'feature1': np.random.normal(loc=0,scale=1,size=set2),
         'feature2': np.random.normal(loc=0,scale=1,size=set2),
          'target': [1]*100  

    
})
#LOC->ORTALAMA
#SCALE->STANDART SAPMA
```

```python
df=pd.concat([df1,df2]).reset_index(drop=True)
```

```python
df.describe()
```

```python
df['target'].value_counts() #burada verirler eşit dağılmamış biz 1'leri artıracağız ya da 0'ları azaltacağız
#UPSAMPLİNG VEYA DOWNSAMPLİNG
```

```python
#UPSAMPLİNG
df_minority=df[df['target']==1]#azınlıklar
df_majority=df[df['target']==0]#çoğunluklar
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsampled=resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42)
```

```python
df_minority_upsampled
```

```python
df_upsampled_1=pd.concat([df_majority,df_minority_upsampled]).reset_index(drop=True) #reset index kullanmazsak daha iyi
sns.scatterplot(data=df_upsampled_1,x='feature1',y='feature2',hue='target')
plt.show()
```

```python
df_upsampled_1.target.value_counts()
```

```python
df_upsampled_1[df_upsampled_1['target']==0]
```

```python
#resample kütüphanesi daha çok aynı dataları kopyalıyor gibi şimdi daha farklı bir kütüphane kullanacağız
#smote kütüphanesi arttırma yapar ama kopyalama yapmaz benzer datalar oluşturur
```

```python
plt.figure(figsize=(10,7),dpi=300)
sns.scatterplot(df,x='feature1',y='feature2',hue='target')
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,OrdinalEncoder
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
cat_cols=df.select_dtypes(exclude=['int64','float64'])
```

```python
for i in cat_cols:
    print(f' {df[i].value_counts()}\n')
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['embarked','embark_town'],inplace=True)
```

```python
df.drop('deck',axis=1,inplace=True)
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
cat_cols
```

```python
df_one_hot=pd.get_dummies(df,dtype='int64',columns=['sex','embark_town'],drop_first=True)
```

```python
df_one_hot
```

```python
df_label=df.copy()
```

```python
labelencoder=LabelEncoder()
```

```python
df_label['sex']=labelencoder.fit_transform(df_label['sex'])
```

```python
df_label
```

```python
df_ordinal=df.copy()
```

```python
classes=['Third','Second','First']
```

```python
ordinalencoder=OrdinalEncoder(categories=[classes])
```

```python
df_ordinal['class']=ordinalencoder.fit_transform(df_ordinal[['class']])
```

```python
df_ordinal
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,5),dpi=300)
df['sex'].value_counts().plot(kind='bar',ax=axes[0],title='original')
df_label['sex'].value_counts().plot(kind='bar',ax=axes[1],title='label encoding',color='red')
df_ordinal['sex'].value_counts().plot(kind='bar',ax=axes[2],title='ordinal',color='black')
```
