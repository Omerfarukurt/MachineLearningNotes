```python
df_apply=pd.read_csv('8-apply_function_data.csv')
```

```python
df_apply
```

```python
#HER SATIRA AYRI AYRI UYGULAMA
def salary_(salary):
    if salary<50000:
        return 'Low'
    elif 50000<=salary<80000:
        return 'Mid'
    else:
        return 'High'
```

```python
df_apply['Salary_sit']=df_apply['Salary'].apply(salary_)
```

```python
df_apply
```

```python
def perf_score(row):
    if row['Experience']>10:
        return row['Performance_Score']+1
    else:
        return row['Performance_Score']
```

```python
df_apply['New_score']=df_apply.apply(perf_score,axis=1)
```

```python
df_apply
```

```python
#DİYELİMKİ BAZI ŞEYLERİN DEĞERİNİ DEĞİŞTİRMEK İSTİYORUZ
df_apply['New_name']=df_apply['Name'].apply(lambda x: x.replace('_',''))
```

```python
#DİYELİMKİ YENİ KOLONU İSTEDİGİMİZ İNDEXE EKLEMEK İSTİYORUZ
df_apply.columns.get_loc('Name')#1-kolonun indexini öğren
df_apply.insert(2,'New_name',df_apply['Name'].apply(lambda x: x.replace('_','')))#sonra insert metoduyla ekleme yap
#ilk parametre kaçıncı index,ikinci sütun ismi,üçüncü ise değerleri alır
```
