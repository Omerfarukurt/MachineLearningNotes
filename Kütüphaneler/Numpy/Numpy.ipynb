```python
#z-score
```

```python
z_score_list=np.array([10,20,30,45,44,33,55,300,321])
```

```python
mean=np.mean(z_score_list)
```

```python
std=np.std(z_score_list)
```

```python
z_score=(z_score-mean)/std
```

```python
z_score #outlier olan verilerin z_scoreleri büyük cıktı tespit etme işlemi aşağıda
```

```python
z_score_list[z_score>1]
```
