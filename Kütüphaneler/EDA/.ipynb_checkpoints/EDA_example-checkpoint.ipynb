```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=sns.load_dataset('penguins')
```

```python
data.head()
```

```python
cat_cols=data.select_dtypes(exclude=['int64','float64'])
```

```python
for i in cat_cols:
    print(f'kolon-> {data[i].value_counts()}')
```

```python
sns.histplot(data['species'])
```

```python
plt.figure(figsize=(10,7),dpi=300)
sns.catplot(data,x='species',kind='bar',y='body_mass_g',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.displot(data,x='body_mass_g',col='species',row='sex',binwidth=20, height=4, facet_kws=dict(margin_titles=True))
plt.show()
```

```python
plt.figure(figsize=(10,7),dpi=300)
sns.catplot(data,x='species',kind='bar',y='bill_length_mm',hue='sex')
plt.show()
```
