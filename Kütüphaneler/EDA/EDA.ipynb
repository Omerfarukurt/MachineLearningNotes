```python
age_list=[10,20,30,30,35,35,37]
weight=[39,65,75,73,80,78,76]
plt.plot(age_list,weight,'y')
plt.xlabel('Yaş')
plt.ylabel('kilo')
plt.show()
#plt.savefig('test.png') grafiği kayıt eder
```

```python
np_array1=np.linspace(0,100,20)
np_arrray2=np_array1
```

```python
plt.plot(np_array1,np_arrray2,"r-o") #rengin yanına parametereler vererek datapointler gösterebiliriz
plt.show()
```

```python
#SUBPLOT KULLANARAK ÇOKLU GRAFİKLER ÇİZİLİR
plt.subplot(1,2,1)#->1 satır 2 kolon 1.grafiği çiziyorum demek
plt.plot(np_array1,np_arrray2,"r-o")
plt.subplot(1,2,2)
plt.plot(np_arrray2,np_array1,"g--")
plt.show()
```

```python
my_figure=plt.figure()
figure_axes=my_figure.add_axes([0.1,0.1,0.4,0.4]) # ilk iki parametre alttan ve üstten boşluğu son ikisi genişlik ve yükseklik
figure_axes.plot(np_array1,np_arrray2/15,'r--')
figure_axes.set_xlabel("X axis")
figure_axes.set_ylabel("Y axis")
figure_axes.set_title("my graph")
figure_axes.set_xlim(left=0.0,right=15.0)
figure_axes.set_ylim(bottom=0,top=15.0)
plt.show()
```

```python
#plt.subplots dediğimizde aslında bize bir fig ve axes veriri
```

```python
(my_fig,my_axes)=plt.subplots()
my_fig=plt.figure(dpi=300)
my_axes.plot(np_array1,np_array1,color='blue')
my_axes.plot(np_array1,np_array1+15,color='black',linewidth=5.0)
my_axes.plot(np_array1,np_array1+30,color='gray',linewidth=3.0,linestyle=':')
my_axes.plot(np_array1,np_array1+45,color='red',linewidth=2.0,linestyle='-.',marker='o',markersize=6,markerfacecolor='white')
my_axes.plot(np_array1,np_array1+60,color='green',linestyle=':',marker='x',markersize=5,markerfacecolor='white')
my_axes.plot(np_array1,np_array1+75,color='purple')

plt.show()
```

```python
df=pd.read_csv('athlete_events.csv')
df.head()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='Age',y='Height',data=df,hue='Sex',style='Medal')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.displot(df,x='Age',col='Medal',row='Sex',binwidth=3, height=3, facet_kws=dict(margin_titles=True))
plt.show()
```
