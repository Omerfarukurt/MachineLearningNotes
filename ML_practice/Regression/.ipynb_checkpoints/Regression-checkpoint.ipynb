```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('1-studyhours.csv')
```

```python
sns.lineplot(data=data,x=data['Study Hours'],y=data['Exam Score'])
```

```python
X=data.drop('Exam Score',axis=1) #veya X=data[['Study Hours']] fazladan bir şey verecekmişiz gibi yazmalıyız
y=data['Exam Score']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=20)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train) #fit ederken std ve mean hesaplanır ona göre sonuç cıkarılır
X_test=scaler.transform(X_test) #data sızmasını engellemek için sadece transform
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression() #n_jobs=-1 verilirse tüm cpular calısır
```

```python
model.fit(X_train,y_train)
```

```python
print('Coefficent:',model.coef_)
print('Intercept:',model.intercept_)
```

```python
#y=73.8+17.09x bizim denklemimiz
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,model.predict(X_train),'r')
```

```python
#eger X'imiz 5 olursa y ne cıkar ama bizim burda 5'i scale edilmiş olarak vermemiz lazım
```

```python
model.predict(scaler.transform([[5]]))
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
acc=[mean_absolute_error,mean_squared_error,r2_score]
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2_score_=r2_score(y_test,y_pred)
acc=[mse,mae,r2_score_]
```
