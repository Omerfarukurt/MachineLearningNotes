```python
X=df.drop('selling_price',axis=1)
y=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
cat_cols=[col for col in df.columns if df[col].dtype=='O']
# cat_cols=df.select_dtypes("object").columns.to_list() aynı şeyi yapar
```

```python
unique_values=df[cat_cols].nunique()
unique_values
```

```python
#seller_type,fuel_type,transmission_type=onehotencoding
#car_name,brand=frequency
```

```python
one_hot_columns=['seller_type','fuel_type','transmission_type']
freq_columns=['car_name','brand','model']
```

```python
for col in freq_columns:
    freq=X_train[col].value_counts()/len(X_train)

    X_train[col+'_freq']=X_train[col].map(freq)
    X_test[col+'_freq']=X_train[col].map(freq)
```

```python
for col in freq_columns:
    X_train.drop(col,axis=1,inplace=True)
    X_test.drop(col,axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(
    transformers=[
        ('onehot',OneHotEncoder(drop='first',handle_unknown='ignore'),one_hot_columns)
        
    ],remainder='passthrough'

    
)
```

```python
X_train=transformer.fit_transform(X_train)
X_test=transformer.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=AdaBoostRegressor()
```

```python
params={
    'n_estimators':[20,50,75,100,150,200],
    'learning_rate':[0.1,0.3,0.5,0.7,1.0,1.5,2.0]
    
}
```

```python
grid=RandomizedSearchCV(estimator=model,param_distributions=params,n_iter=10,n_jobs=4)
```

```python
grid.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error,mean_absolute_error
```

```python
pred=grid.predict(X_test)
```
